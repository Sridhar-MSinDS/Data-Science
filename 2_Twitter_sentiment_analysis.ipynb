**Import necesaary libraries**

```python
import re
from ntscraper import Nitter
from collections import Counter
import matplotlib.pyplot as plt
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from scipy.special import softmax
```

**Tweets Scraping**

```python
# Assuming tweets is a dictionary with 'tweets' as a list of tweet information
scraper = Nitter(log_level=1,skip_instance_check=False)
tweets = scraper.get_tweets('data',mode='term',number=100)
```

**Load Model and Tokenizer**

```python
# Load model and tokenizer
roberta = "cardiffnlp/twitter-roberta-base-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)
```

**Sentiment Analysis**

```python
# Check if tweets were successfully fetched
if 'tweets' in tweets and tweets['tweets']:
    # Load model and tokenizer once
    roberta = "cardiffnlp/twitter-roberta-base-sentiment"
    model = AutoModelForSequenceClassification.from_pretrained(roberta)
    tokenizer = AutoTokenizer.from_pretrained(roberta)
    labels = ['Negative', 'Neutral', 'Positive']

    # Process each tweet
    for tweet_info in tweets['tweets']:
        # Preprocess tweet text
        tweet_words = []
        for word in tweet_info['text'].split():  # Ensure you're splitting the 'text' of the tweet
            if word.startswith('@') and len(word) > 1:
                word = '@user'
            elif word.startswith('http'):
                word = "http"
            tweet_words.append(word)
        tweet_proc = " ".join(tweet_words)

        # Perform sentiment analysis
        encoded_tweet = tokenizer(tweet_proc, return_tensors='pt')
        output = model(**encoded_tweet)
        scores = output[0][0].detach().numpy()
        scores = softmax(scores)

        # Print the sentiment scores along with the tweet
        print(f"Processed Tweet: {tweet_proc}")
        print("Sentiment Scores:")
        for label, score in zip(labels, scores):
            print(f"{label}: {score:.4f}")
        print('\n')
else:
    print("No tweets found or error in fetching tweets.")
```
